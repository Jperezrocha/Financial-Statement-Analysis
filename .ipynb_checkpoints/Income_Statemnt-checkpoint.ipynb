```python
import requests
import datetime as dt
import pandas as pd
```

```python
ticker = "AAPL"
api_key = "ec2d6f3b0cc9395fbce02f1316669229"
years = 10
```

```python
url = (f"https://financialmodelingprep.com/api/v3/income-statement/{ticker}?limit={years}&apikey={api_key}")
r = requests.get(url)
income_statement = r.json()

print(income_statement)
```

```python
dates = list(reversed([income_statement[date]['date']for date in range(len(income_statement))]))
revenues = list(reversed([income_statement[i]['revenue']for i in range(len(income_statement))]))

print(dates)
print(revenues)
```

```python
df_income_statement = pd.DataFrame(income_statement)
print(df_income_statement)
```
